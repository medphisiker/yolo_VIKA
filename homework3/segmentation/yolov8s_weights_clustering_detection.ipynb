```python
def parse_annotation(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    image_name = root.find("filename").text
    image_path = os.path.join(path_images, image_name)

    boxes = []
    classes = []
    for obj in root.iter("object"):
        cls = obj.find("name").text
        if cls in classes_ids:
          classes.append(cls)

          bbox = obj.find("bndbox")
          xmin = float(bbox.find("xmin").text)
          ymin = float(bbox.find("ymin").text)
          xmax = float(bbox.find("xmax").text)
          ymax = float(bbox.find("ymax").text)
          boxes.append([xmin, ymin, xmax, ymax])

    class_ids = [
        list(class_mapping.keys())[list(class_mapping.values()).index(cls)]
        for cls in classes
    ]
    return image_path, boxes, class_ids


image_paths = []
bbox = []
classes = []
for xml_file in tqdm(xml_files):
    image_path, boxes, class_ids = parse_annotation(xml_file)
    if class_ids:
      image_paths.append(image_path)
      bbox.append(boxes)
      classes.append(class_ids)
```

```python
def load_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    return image


def load_dataset(image_path, classes, bbox):
    # Read Image
    image = load_image(image_path)
    bounding_boxes = {
        "classes": tf.cast(classes, dtype=tf.float32),
        "boxes": bbox,
    }
    return {"images": tf.cast(image, tf.float32), "bounding_boxes": bounding_boxes}
```

```python
augmenter = keras.Sequential(
    layers=[
        keras_cv.layers.RandomFlip(mode="horizontal", bounding_box_format="xyxy"),
        keras_cv.layers.RandomShear(
            x_factor=0.2, y_factor=0.2, bounding_box_format="xyxy"
        ),
        keras_cv.layers.JitteredResize(
            target_size=(640, 640), scale_factor=(0.75, 1.3), bounding_box_format="xyxy"
        ),
    ]
)
```

```python
BATCH_SIZE = 4
```

```python
train_ds = train_data.map(load_dataset, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.shuffle(BATCH_SIZE * 4)
train_ds = train_ds.ragged_batch(BATCH_SIZE, drop_remainder=True)
train_ds = train_ds.map(augmenter, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
resizing = keras_cv.layers.JitteredResize(
    target_size=(640, 640),
    scale_factor=(0.75, 1.3),
    bounding_box_format="xyxy",
)

val_ds = val_data.map(load_dataset, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.shuffle(BATCH_SIZE * 4)
val_ds = val_ds.ragged_batch(BATCH_SIZE, drop_remainder=True)
val_ds = val_ds.map(resizing, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
cluster_num = 8
```

```python
import tensorflow_model_optimization as tfmot

cluster_weights = tfmot.clustering.keras.cluster_weights
CentroidInitialization = tfmot.clustering.keras.CentroidInitialization

clustering_params = {
  'number_of_clusters': cluster_num,
  'cluster_centroids_init': CentroidInitialization.KMEANS_PLUS_PLUS
}
```

```python
# https://www.tensorflow.org/model_optimization/guide/clustering/clustering_comprehensive_guide

# Helper function uses `cluster_weights` to make only
# the Dense layers train with clustering
def apply_clustering_to_dense(layer):
    if isinstance(layer, tf.keras.layers.Conv2D):
        if layer.trainable_weights:
            try:
                print(f'Кластеризую веса у слоя - {layer.name}')
                return cluster_weights(layer, **clustering_params)
            except:
                return layer

    return layer
```

```python
# Use `tf.keras.models.clone_model` to apply `apply_clustering_to_dense`
# to the layers of the model.
clustered_model = tf.keras.models.clone_model(
    backbone,
    clone_function=apply_clustering_to_dense,
)
```

```python
clustered_model.summary()
```

```python
import numpy as np

def print_cluster_layers_weights(clustered_model, cluster_num):
    # проверим уникальные веса
    for i in clustered_model.weights:
        uniq_values = np.unique(i.numpy())
        if len(uniq_values) == 8:
            print(i.name, uniq_values)
```

```python
print_cluster_layers_weights(clustered_model, cluster_num)
```

```python
# удаляем дополнительные слои
# https://www.tensorflow.org/model_optimization/guide/clustering/clustering_comprehensive_guide#deployment
final_model = tfmot.clustering.keras.strip_clustering(clustered_model)
```

```python
for name in сopy_layers:
    weights = final_model.get_layer(name).get_weights()

    if len(weights) > 1:
        kernel = weights[0]
        bias = weights[1]
        backbone.get_layer(name).set_weights([kernel, bias])
    else:
        kernel = weights[0]
        backbone.get_layer(name).set_weights([kernel])
```
